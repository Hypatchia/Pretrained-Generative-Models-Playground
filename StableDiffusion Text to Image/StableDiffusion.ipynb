```python
# imports
import time
import keras_cv
import keras
import matplotlib.pyplot as plt
```

```python
# Instantiate the stable diffusion model from keras_cv
model = keras_cv.models.StableDiffusion(img_width=512,img_height=512,jit_compile=False)
```

```python
# Generate batch_size=3 images using text prompt
images = model.text_to_image("Photograph of a cat wearing a halloween custome",batch_size=3)
```

```python
# Define a function to plot images
def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")
```

```python
# Plot result of first prompt
plot_images(images)
```

```python
# Second prompt
images = model.text_to_image("Photograph of a pink panther wearing a suit",batch_size=3)
```

```python
# plot result of second prompt
plot_images(images)
```
